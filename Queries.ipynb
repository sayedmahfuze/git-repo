   - **Answer:**

     class LRUCache:
         def __init__(self, capacity):
             self.cache = OrderedDict()
             self.capacity = capacity

         def get(self, key):
             if key not in self.cache:
                 return -1
             else:
                 self.cache.move_to_end(key)
                 return self.cache[key]

         def put(self, key, value):
             if key in self.cache:
                 self.cache.move_to_end(key)

         dp = [[0] * (n + 1) for _ in range(m + 1)]

5. **Implement an LFU Cache:**
   - **Question:** Design and implement a Least Frequently Used (LFU) cache. Implement `get` and `put` methods to manage cache entries and track their usage frequency.
   - **Answer:**
     ```python
     from collections import defaultdict, OrderedDict

     class LFUCache:
         def __init__(self, capacity):
             self.capacity = capacity
             self.cache = {}
             self.freq = defaultdict(OrderedDict)
             self.min_freq = 0

         def get(self, key):
             if key not in self.cache:
                 return -1
             value, freq = self.cache[key]
             self.freq[freq].pop(key)
             if not self.freq[freq]:
                 del self.freq[freq]
                 if self.min_freq == freq:
                     self.min_freq += 1
             self.cache[key] = (value, freq + 1)
             self.freq[freq + 1][key] = None
             return value

         def put(self, key, value):
             if self.capacity <= 0:
                 return
             if key in self.cache:
                 self.get(key)
                 self.cache[key] = (value, self.cache[key][1])
             else:
                 if len(self.cache) >= self.capacity:
                     old_key, _ = self.freq[self.min_freq].popitem(last=False)
                     del self.cache[old_key]
                 self.cache[key] = (value, 1)
                 self.freq[1][key] = None
                 self.min_freq = 1
     ```

6. **Find the Shortest Path in a Grid with Obstacles:**
   - **Question:** Given a 2D grid with obstacles (represented by `1`), find the shortest path from the top-left corner to the bottom-right corner. Movement is allowed in four directions: up, down, left, and right.
   - **Answer:**
     ```python
     from collections import deque

     def shortest_path(grid):
         rows, cols = len(grid), len(grid[0])
         directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
         queue = deque([(0, 0, 0)])  # (row, col, distance)
         visited = set((0, 0))

         while queue:
             row, col, dist = queue.popleft()
             if (row, col) == (rows - 1, cols - 1):
                 return dist
             for dr, dc in directions:
                 new_row, new_col = row + dr, col + dc
                 if (0 <= new_row < rows and 0 <= new_col < cols and
                     grid[new_row][new_col] == 0 and (new_row, new_col) not in visited):
                     queue.append((new_row, new_col, dist + 1))
                     visited.add((new_row, new_col))
         return -1
     ```

7. **Calculate the Maximum Width of a Binary Tree:**

   - **Answer:**
     ```python
     from collections import deque

     class TreeNode:
         def __init__(self, value=0, left=None, right=None):
             self.value = value
             self.left = left
             self.right = right

     def width_of_binary_tree(root):
         if not root:
             return 0

         max_width = 0
         queue = deque([(root, 0)]

         while queue:
             level_length = len(queue)
             _, first_index = queue[0]
             for _ in range(level_length):
                 node, index = queue.popleft()
                 if node.left:
                     queue.append((node.left, 2 * index))
                 if node.right:
                     queue.append((node.right, 2 * index + 1))
             max_width = max(max_width, index - first_index + 1)

         return max_width
     ```

Object-Oriented Programming (OOP) in Python

Object-Oriented Programming is a programming paradigm that uses objects and classes to
structure software in a way that models real-world entities and their interactions.

Core Concepts of OOP

1. Class: A blueprint for creating objects. It defines a set of attributes and methods that the created objects (instances) will have.

2. Object: An instance of a class. It is a real-world entity with state (attributes) and behavior (methods)

3. Encapsulation: The bundling of data (attributes) and methods that operate on the data into a single unit or class. It restruct direct access to some of the object's components.

4. Iheritance: A mechanism where a new class derives properties and behaviors from an existing class.

5. Polymorphism: The ability to present the same interface for different underlying data types. for Example, a method can process objects differently based on their class.

6. Abstraction: Hiding the complex implememtation details and showing only the essential features of an object.

````python
# Abstraction focuses on hiding the complexity and only showing the essential features of an object

from abc import ABC, abstractmethod


class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

  @abstractmethod
  def perimeter(self):
    pass

class Rectangle(Shape):
  def __init__(self, width, height):
    self.width = width
    self.height = height

  def area(self):
    return self.width * self.height

  def perimeter(self):
    return 2 * (self.width + self.height)


rect = Rectangle(4, 7)
print(rect.area())
print(rect.perimeter())





````
